```python
from scipy import sparse
```

```python
eye=np.eye(4)
print(eye)
print("numpy array:\n{}".format(eye))
```

```python
sparse_matrix=sparse.csr_matrix(eye)
print("\n SCipy sparse CSR matrix for this is:\n{}".format(sparse_matrix))
```
